```python
#Name: Manish Yegurla
#WSU_ID: Z488Y969

import numpy as np
import sys
f = open("output.txt", 'w')
sys.stdout = f


list_old = []
def number_of_attacks(queen, x, y):
  n = 0
  for z in range(0,8):
    if queen == grid[x,z]:
      n += 1
    if abs(x - row1[z]) == abs(y - column[z]):
      n += 1
  return n-2

#row1 is used to change row values just for the functional use in change_positions()
row1 = [0,0,0,0,0,0,0,0]
def create_rows_columns(list):
  for y in range(0,8):
    x = int(list[y])
    grid[x-1][y] = 1
    row.append(x-1)
    column.append(y)
  for i in range(8):  
    row1[i] = row[i]

na = []
def max_number_of_attacks():
  for i in range(0,8):
      na.append(number_of_attacks(grid[row[i],column[i]], row[i], column[i]))
  n1 = max(na)

queens_max_attacks = []
def queens_with_max_attacks():
  if max(na) == 0:
      print("Success")
      SystemExit
  else:
    for i in range(len(na)):
      if na[i] == max(na):
        queens_max_attacks.append(i)

def change_positions(queens_max_attacks):
  n1 = max(na)
  #Loop through all the queens having same max attacks
  for f in range(len(queens_max_attacks)):
    grid[row[queens_max_attacks[f]],column[queens_max_attacks[f]]] = 0
    n3 = []
    #Loop used to take number of attacks for a queen in all the rows and append them to n3
    for i in range(0,8):
      grid[i,column[queens_max_attacks[f]]] = 1
      new_queen = grid[i,column[queens_max_attacks[f]]]
      row1[queens_max_attacks[f]] = i
      #number of attacks for a queen chosen in the state
      n3.append(number_of_attacks(new_queen,i,column[queens_max_attacks[f]])) 
      grid[i,column[queens_max_attacks[f]]] = 0
    m = min(n3) #minimum number of attacks for the chosen queen
    if m < n1:
      grid[n3.index(m),column[queens_max_attacks[f]]] = 1
      list_old = list.copy()
      list[queens_max_attacks[f]] = n3.index(m)+1
      print(*list)
      na.clear()
      row.clear()
      column.clear()
      create_rows_columns(list)
      max_number_of_attacks()
      if max(na) == 0:
        print("success")
        print("")
        break
      else:
        queens_max_attacks.clear()
        queens_with_max_attacks()
        if m >= n1:
          print("min of n3 is less than n1")
          break
        elif n3 == []:
          print("n3 is empty")
          break
        else:
            if list_old != list:
              change_positions(queens_max_attacks)
              break
    else:
      grid[row[queens_max_attacks[f]],column[queens_max_attacks[f]]] = 1
      print("Failure")
      print("")
      break

number_of_states = 0
def start():
  flag = 0
  number_of_states = int(input("Enter total number of states: "))
  lists = np.zeros((number_of_states,8))
  for k in range(number_of_states):
    list = input("Enter 8 numbers for state {}: ".format(k+1))
    queens_of_one_state = list.split()
    if len(list) != 15:
      print("Length of list is not 15")
      break
    elif len(list) == 15:
      if len(queens_of_one_state) != 8:
        print("Not enough values")
        break
      elif len(queens_of_one_state) == 8:
        for i in range(0,8):
          if not 1 <= int(queens_of_one_state[i]) <=8:
            print("Enter values between 1-8!!")
            flag += 1
            break
    if flag == 0:
      for j in range(8):
        lists[k,j] = queens_of_one_state[j]
    else:
      break
  return lists

#Takes all the input states as a grid
all_states_in_a_grid = start()
min_value_of_all_states = []
stop = 0

#Checks if any state is given wrong input
for i in range(len(all_states_in_a_grid)):
  min_value_of_all_states.append(min(all_states_in_a_grid[i]))
  if min_value_of_all_states[i] == 0:
    stop += 1

#Initializing row and column variables to hold the values of the queens list
row = []
column = []
n2 = 0
n3 = []
list = []

if stop == 0:
  for i in range(len(all_states_in_a_grid)):
    list.clear()
    n3.clear()
    na.clear()
    row.clear()
    queens_max_attacks.clear()
    grid = np.zeros((8,8))

    risk = [] 
    for g in range(8):
      risk.append(int(all_states_in_a_grid[i,g]))
    list = risk
    list_old = list
    print("")
    print(*list)

    #Calling the functions
    create_rows_columns(list)

    #Calls function to se "na" which is max number of attacks for a state
    max_number_of_attacks()
    n1 = max(na) #maximum number of attacks for a particular state

    #This function is used to store all the queens having max attacks
    queens_with_max_attacks()
    change_positions(queens_max_attacks)
```
